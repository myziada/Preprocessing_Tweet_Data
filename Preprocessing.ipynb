```python
import pandas as pd
import re

# Read the CSV file into a DataFrame
df = pd.read_csv('output1.csv')

# Define a function to remove usernames from a tweet
def remove_username(tweet):
    if isinstance(tweet, str):  # Memeriksa jika nilai tweet adalah string
        cleaned_tweet = re.sub(r'@\w+', '', tweet)
        return cleaned_tweet
    else:
        return tweet

# Apply the function to the 'content' column
df['Text'] = df['content'].apply(remove_username)

# Save the updated DataFrame to a new CSV file
df.to_csv('cleaned_tweets.csv', index=False)
```

```python
import re
import pandas as pd

def remove_emoticons(text):
    emoticon_pattern = re.compile("["
                                  u"\U0001F600-\U0001F64F"  # emoticons
                                  u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                                  u"\U0001F680-\U0001F6FF"  # transport & map symbols
                                  u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                                  u"\U00002500-\U00002BEF"  # chinese char
                                  u"\U00002702-\U000027B0"
                                  u"\U00002702-\U000027B0"
                                  u"\U000024C2-\U0001F251"
                                  u"\U0001f926-\U0001f937"
                                  u"\U00010000-\U0010ffff"
                                  u"\u2640-\u2642"
                                  u"\u2600-\u2B55"
                                  u"\u200d"
                                  u"\u23cf"
                                  u"\u23e9"
                                  u"\u231a"
                                  u"\ufe0f"  # dingbats
                                  u"\u3030"
                                  "]+", flags=re.UNICODE)
    return emoticon_pattern.sub(r'', text)

# Read the CSV file
data = pd.read_csv('modified_file3.csv')
data['Text'] = data['Text'].astype(str)

# Remove emoticons from each text entry
data['Text'] = data['Text'].apply(remove_emoticons)

# Save the modified data to a new CSV file
data.to_csv('output5.csv', index=False)
data
```

```python
import pandas as pd

# Read the CSV file
data = pd.read_csv('output8.csv')

# Function to add a space before "Anies" or "anies" if not already present
def add_space_before_anies(text):
    # Check if the value is NaN
    if pd.isna(text):
        return text

    # Check if "Anies" or "anies" is present without a space before it
    if "Anies" in text and not text.startswith("Anies "):
        text = text.replace("Anies", " Anies")
    if "anies" in text and not text.startswith("anies "):
        text = text.replace("anies", " anies")
    return text

# Apply the function to the 'text' column, handling NaN values
data['Text'] = data['Text'].apply(lambda x: add_space_before_anies(x) if isinstance(x, str) else x)

# Drop rows with NaN values in the 'text' column
data = data.dropna(subset=['Text'])

# Save the modified data to a new CSV file
data.to_csv('modified_file9.csv', index=False)
data
```

```python
import pandas as pd
import re

def remove_duplicate_letters(text):
    # Find words with duplicate letters using a regular expression pattern
    pattern = r'(\w)\1+'
    new_text = re.sub(pattern, r'\1', text)
    return new_text if isinstance(text, str) else ''

# Read the CSV file
data = pd.read_csv('output_output_lowercase.csv.csv')

# Convert the 'Text' column to string type
# data['Text'] = data['Text'].astype(str)

# Apply the remove_duplicate_letters function to the 'Text' column
data['text'] = data['text'].apply(remove_duplicate_letters)

# Remove rows with empty text entries
data = data[data['text'].str.strip() != '']

# Save the modified data to a new CSV file
data.to_csv('outputput.csv', index=False)
data
```

```python
import csv

def hapus_duplikat(teks):
    kata = teks.split()
    kata_unik = list(set(kata))
    teks_baru = ' '.join(kata_unik)
    return teks_baru

nama_file_input = "output_output_lowercase.csv.csv"
nama_file_output = "outputput.csv"

# Membaca data dari file CSV
data = []
with open(nama_file_input, 'r') as file_input:
    reader = csv.reader(file_input)
    header = next(reader)  # Membaca baris header
    for row in reader:
        data.append(row)

# Menghapus kata-kata duplikat dalam kolom teks
for row in data:
    teks = row[0]  # Misalnya, teks berada di kolom kedua (indeks 1)
    teks_tanpa_duplikat = hapus_duplikat(teks)
    row[0] = teks_tanpa_duplikat

# Menyimpan data ke file CSV
with open(nama_file_output, 'w', newline='') as file_output:
    writer = csv.writer(file_output)
    writer.writerow(header)
    writer.writerows(data)

print("Data telah diproses dan disimpan dalam file", nama_file_output)
```

```python
import csv
import string
import re

input_file = 'outputput.csv'
output_file = '!output_removePunctuationAndNumber.csv'

# Function to remove punctuation and numbers from a string
def remove_punctuation_and_numbers(text):
    # Remove punctuation
    translator = str.maketrans('', '', string.punctuation)
    text = text.translate(translator)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    return text

# Read the CSV file
with open(input_file, 'r') as file:
    reader = csv.reader(file)
    data = list(reader)

# Remove punctuation and numbers from each cell
modified_data = [[remove_punctuation_and_numbers(cell) for cell in row] for row in data]

# Write the modified data to a new CSV file
with open(output_file, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(modified_data)

print("Punctuation and numbers removed from CSV file and saved to output CSV file.")
df = pd.read_csv(output_file)
df
```

```python
import pandas as pd

# Read the CSV file
data = pd.read_csv('output161.csv')

# Function to add a space before "Anies" or "anies" if not already present
def add_space_before_anies(text):
    # Check if the value is NaN
    if pd.isna(text):
        return text

    # Check if "Baswedan" or "baswedan" is present without a space before it
    if "Baswedan" in text and not text.startswith("Baswedan "):
        text = text.replace("Baswedan", " Baswedan")
    if "baswedan" in text and not text.startswith("baswedan "):
        text = text.replace("baswedan", " baswedan")
    return text

# Apply the function to the 'text' column, handling NaN values
data['text'] = data['text'].apply(lambda x: add_space_before_anies(x) if isinstance(x, str) else x)

# Drop rows with NaN values in the 'text' column
data = data.dropna(subset=['text'])

# Save the modified data to a new CSV file
data.to_csv('output17.csv', index=False)
data
```

To normalize Indonesian tweet data, you can apply various text normalization techniques. Here are a few common steps you can take:

Convert text to lowercase: Convert all text to lowercase to ensure consistent casing. This helps to treat uppercase and lowercase versions of the same word as the same.

Remove special characters: Remove any special characters, such as punctuation marks or symbols, that may not contribute significantly to the meaning of the text.

Remove URLs: Remove any URLs or links present in the tweet text. These may not add value to the analysis and can be safely removed.

Tokenization: Split the tweet text into individual tokens or words. This step helps to process and analyze the text at a more granular level.

Remove stop words: Remove common words, known as stop words, such as "and," "the," "is," etc. These words typically occur frequently in the text and do not provide much meaning for analysis.

Stemming or lemmatization: Apply stemming or lemmatization techniques to reduce words to their root form. This step helps to consolidate variations of the same word and reduces the overall vocabulary.

```python
import csv
import re

def remove_excess_letters(text):
    cleaned_text = re.sub(r'(\w)\1+', r'\1', text)
    return cleaned_text

input_file = 'output21.csv'
output_file = 'output3.csv'

# Open the input CSV file and create the output CSV file
with open(input_file, 'r', newline='') as input_csv, open(output_file, 'w', newline='') as output_csv:
    reader = csv.reader(input_csv)
    writer = csv.writer(output_csv)

    # Iterate over each row in the input CSV file
    for row in reader:
        modified_row = []

        # Process each column in the row
        for column in row:
            # Remove excess letters from the column text
            cleaned_text = remove_excess_letters(column)
            modified_row.append(cleaned_text)

        writer.writerow(modified_row)
```

```python
import csv
import re

def remove_duplicate_adjacent_words(text):
    pattern = r'\b(\w+)(\s+\1\b)+'
    cleaned_text = re.sub(pattern, r'\1', text)
    return cleaned_text

input_file = 'output3.csv'
output_file = 'output4.csv'

# Open the input CSV file and create the output CSV file
with open(input_file, 'r', newline='') as input_csv, open(output_file, 'w', newline='') as output_csv:
    reader = csv.reader(input_csv)
    writer = csv.writer(output_csv)

    # Iterate over each row in the input CSV file
    for row in reader:
        modified_row = []

        # Process each column in the row
        for column in row:
            # Remove duplicate adjacent words from the column text
            cleaned_text = remove_duplicate_adjacent_words(column)
            modified_row.append(cleaned_text)

        writer.writerow(modified_row)
```

```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download stopwords for Indonesian
nltk.download('stopwords')
nltk.download('punkt')

# Load stopwords for Indonesian
stopwords_ind = set(stopwords.words('indonesian'))

def preprocess_text(text):
    # Convert text to string
    text = str(text)

    # Remove repeated characters
    processed_text = re.sub(r'(\w)\1+', r'\1', text)

    # Normalize word endings
    processed_text = re.sub(r'(\w)(\w)\2(\w)', r'\1\2\3', processed_text)

    # Tokenize the text into words
    words = word_tokenize(processed_text)

    # Remove stopwords
    filtered_words = [word for word in words if word.lower() not in stopwords_ind]

    # Join the filtered words back into a string
    preprocessed_text = ' '.join(filtered_words)

    return preprocessed_text

# Read the CSV file into a DataFrame
df = pd.read_csv('output4.csv')

# Specify the column containing the text
text_column = 'text'

# Create a new column for storing the processed text
processed_column = 'processed_text'
df[processed_column] = df[text_column].apply(preprocess_text)

# Save the preprocessed DataFrame to a new CSV file
df.to_csv('output_stopwordForFix.csv', index=False)
df
```
